```python
import requests
api = "https://apidadosabertos.saude.gov.br/vigilancia-e-meio-ambiente/sistema-de-informacao-sobre-nascidos-vivos?limit=1000"
```

```python
response = requests.get(api)
content = response.json()
```

```python
content['sinasc']
```

```python
full_content = []
for i in range (2):
    content = requests.get(f"{api}&offset={i}").json()
    for item in content['sinasc']:
        if 'codufnatu' in item.keys():
            full_content.append(item)
```
